```python
#ploting density plot
def density_plot(feature,class_):
    """this function plots density curves for given feature and the class"""
    sns.set_style(style="whitegrid")
    sns.FacetGrid(data=dataset,hue=class_,size=7).map(sns.distplot,feature).add_legend()
    plt.show()
```

```python
density_plot("Age","Survival status")
```

```python
density_plot("Year of operation","Survival status")
```

```python
density_plot("axillary nodes","Survival status")
```

# observaton
### The curves in age and year of operation are over lapping with larger area
### whereas in axillary nodes the curvrs are indicating that there are large number of people with satus one with nodes ranging from -1 to 5
